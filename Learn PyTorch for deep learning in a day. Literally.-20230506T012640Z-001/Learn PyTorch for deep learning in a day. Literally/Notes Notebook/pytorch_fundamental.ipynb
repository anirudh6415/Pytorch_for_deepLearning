###matrix multiplication 
####2 types of multipliction
1. element wise 
2. matrix multiplication (dot product)

there are two main rules that performing matrix multiplication needs to satisfy :

1. the **inner dimension** must match:
* `(3,2) @ (3,2) ` wont work
* `(2,3) @(3,2) ` will work
* `(3,2) @ (2 ,3)` will work

2. the resulting matrix has the shape of the ** outer dimension **
* `(2,3) @ (3,2)` ->(2,2)
* `(3,2) @ (2,3)` ->(3,3)


```python
#element wise 
print(tensor,"*",tensor)
print(f"Equals:{tensor*tensor}")
```

```python
#matrix mul
torch.matmul(tensor,tensor)
tensor.shape 
```

## reshaping , Stacking , Squeezing and Unsqueezing

* Reshaping  - Reshapes an input tensor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* stacking  - combine multiple tensor on top of each other (vstack) or side by side (hstack)
* squeeze - removes all '1' dimensions from the tensor
* unsqueeze - add a '1' dimension to  a target tensor
* permute - return a view of the input with dimension permuted

```python
import torch 
x = torch.arange(1.,9.)
x,x.shape
```

```python
#add an extra dimension
x_reshape = x.reshape(1,8)
x_reshape, x_reshape.shape
```

```python
#change the view 
z = x.view(1,8)
z,z.shape
```

```python
z[:,0] =5
z,x
```

```python
# stack tensor on top of each other
X_stacked = torch.stack([x,x,x,x],dim = 0)
X_stacked
```

```python
#squeezetensor
y = torch.squeeze(x_reshape,0)
xsqueezed  = x_reshape.squeeze()
y , y.shape,xsqueezed,xsqueezed.shape
```

```python
#unsqueeze 
print(f"pervious target : {xsqueezed}")
print(f"pervious shape : {xsqueezed.shape }")
xunsqueezed = xsqueezed.unsqueeze(dim = 0)
print(f"new target : {xunsqueezed}")
print(f"new shape : {xunsqueezed.shape }")
```

```python
#torch.permute - rearrages the target tensor with specified order
x_original = torch.rand(size = (224,224,3)) #[height , width , color channels]\

#permute the original tensor to rearrage the axis order
xpermuted = x_original.permute(2,0,1)
print("pervious shape",x_original.shape)
print("new shape", xpermuted.shape) #[color channels,height , width]\
```

```python
x_original[0,0,0] =74
xpermuted[0,0,0]
```

## pytorch tensors and numpy
numpy is popluar scientific python numerical computing library.

and because of this , pytorch has functionality to interact with it.

* Data in numpy , want in pytorch tensor -> `torch.from_numpy(ndarray)`

* pytorch tensor -> numpy  -> `torch.tensor.numpy`




```python
import torch
import numpy as np

array = np.arange(1.0,8.0)
tensor = torch.from_numpy(array) #warning when converting numpy to pytorch 
                              # pytorch reflects numoys default datatype of float64 unless specified otherwise
                              # to change to float32 torch.from_numpy(ndarray).type(torch.float32)

array,tensor
```

```python
#change the value of array it doesnt change in tensor
array = array +1
array, tensor
```

## Reproducbility (trying to take random out of random)
in short how a neural network learns :
` start with random number -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> again ... `

to reduce the randomness in neural network and pytorch comes the concept of a **random seed**.
Essentially what the random seeds does is  "flavor" the randomness.


```python
import torch
randomA = torch.rand(3,4)
randomB = torch.rand(3,4)

print(randomA)
print(randomB)
print(randomA == randomB)
```

```python
# lets's make some random but reproducible tensor
import torch

Random_seed = 42

torch.manual_seed(Random_seed)
randomc = torch.rand(3,4)

torch.manual_seed(Random_seed)
randomd = torch.rand(3,4)

print(randomc)
print(randomd)
print(randomc == randomd)
```
